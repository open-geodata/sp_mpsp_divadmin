# Áreas Regionais


```python
import open_geodata as geo
import pandas as pd

from sp_mpsp_divadmin.paths import input_path_tab, output_path_tabs
```

<br>

## Nomes dos Municípios

Lê o arquivo csv com o nome dos municípios


```python
df_mun = geo.load_dataset(db='sp', name='tab.municipio_nome')

df_mun.info()
df_mun.head()
```

<br>

## IDs vinculando Municípios com Áreas Regionais


```python
# Dataframe
df = pd.read_csv(
    input_path_tab / 'tab_ids_municipio.csv',
    usecols=['id_municipio', 'id_ar'],
)

# Results
df.info()
df.head()
```

```python
# Merge Municípios
df = pd.merge(
    df_mun,
    df,
    how='left',
    left_on='id_municipio',
    right_on='id_municipio',
)

# Results
df.info()
df.head()
```

<br>

## Município Áreas Regionais


```python
# Dataframe GAEMAs
df_ar = pd.read_csv(input_path_tab / 'tab_arearegional.csv')
df_ar = df_ar.add_prefix('ar_')

# Results
df_ar.info()
df_ar.head()
```

```python
# Merge GAEMAs
df = pd.merge(
    df,
    df_ar,
    how='left',
    left_on='id_ar',
    right_on='ar_id',
)

# Drop
df.drop(['ar_id'], axis=1, inplace=True, errors='ignore')

# Save
df.to_csv(
    output_path_tabs / 'tab_municipio_ar.csv',
    index=False,
)

# Results
df.info()
df.head()
```
